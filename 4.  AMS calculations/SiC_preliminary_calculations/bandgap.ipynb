```python
import matplotlib.pyplot as plt
import numpy as np
import pandas   as pd

#following from https://lampz.tugraz.at/~hadley/ss1/bands/bandstructures/sic_bands.html
df = pd.read_csv('bandgap.csv', delim_whitespace=True)
```

```python
import itertools
X =[]
Y =[]
for t in range(len(df)):
    X.append(df['A'][t])
    Y.append(df['B'][t])
    
left=[]
right=[]
left2=[]
right2=[]
for i, j in zip(X, Y):
    if i < 1.0:
        left.append(i)
        left2.append(j)
    else:
        right.append(i)
        right2.append(j)
plt.figure(figsize=(10,10))
plt.scatter(X,Y,s=0.5)
position =[0.0, 0.5777, 0.91, 1.0, 1.3333, 1.9233, 2.03]
labels = ['\u0393', 'M', 'K', 'H', 'L', 'A', '\u0393']
plt.xticks(position, labels)
plt.title("Hexagonal SiC")
plt.ylabel("Energy (eV)")
plt.show()

plt.figure(figsize=(5,5))
plt.scatter(right,right2,s=0.5)
plt.show()

plt.figure(figsize=(5,5))
plt.scatter(left,left2,s=0.5)
plt.show()
```

```python
reversed = []
reversed2 = []

for t,i in zip(range(len(right)), range(len(right2))):
    reversed.append(right[t]-1)
    reversed2.append(right2[i])
for l1, l2 in zip(left, left2):
    l1+=1.03
    reversed.append(l1)
    reversed2.append(l2)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(reversed, reversed2, s=0.5)

position =[1.03, 1.5777, 1.91, 2.03, 0.3333, 0.9233, 0.0]
labels = ['\u0393', 'M', 'K', 'H', 'L', 'A', 'H']
plt.xticks(position, labels) 
plt.ylabel("Energy (eV)")
plt.show()
```

```python
leftbound = 0.1
rightbound = 1.0
ticks = rightbound*10
first_curve = []

for t, num in zip(range(len(X)),X):
    if leftbound < num < rightbound:
        if (X[t] == num):
            first_curve.append(t)

first_curve_positions = []
for pos in first_curve:
    first_curve_positions.append(X[pos])

positions_energy_nonzero = []
first_curve_energies = []
i=0
for pos in first_curve:
    energy = Y[pos]
    if 14.8 > energy > 10.0:
        first_curve_energies.append(energy)
        positions_energy_nonzero.append(i)
    else: 
        energy = 0.0
        first_curve_energies.append(energy)
    i=i+1

plt.figure(figsize=(10,10))
plt.scatter(first_curve_positions,first_curve_energies,s=0.5)
plt.xticks(np.linspace(leftbound, rightbound, int(ticks)))
plt.ylabel("Energy (eV)")
plt.title("Left 'valley'")
plt.grid()
plt.show()
```

```python
#Remove zeros from energies
last_energies = first_curve_energies

while last_energies.index(0.0) == False:
    last_energies.remove(0.0)
while last_energies.index(0.0) != 0:
    last_energies.remove(0.0)
```

```python
positions_uniques = (0.1013, 0.1114, 0.1215, 0.1317, 0.1418, 0.1519, 0.1621, 0.1722, 0.1823, 0.1925, 0.2026, 0.2127, 0.2228, 0.233, 0.2431, 0.2532, 0.2634, 0.2735, 0.2836, 0.2937, 0.3039, 0.314, 0.3241, 0.3343, 0.3444, 0.3545, 0.3646, 0.3748, 0.3849, 0.395, 0.4052, 0.4153, 0.4254, 0.4355, 0.4457, 0.4558, 0.4659, 0.4761, 0.4862, 0.4963, 0.5064, 0.5166, 0.5267, 0.5368, 0.547, 0.5571, 0.5672, 0.5774, 0.5875, 0.5976, 0.6077, 0.6178, 0.6279, 0.638, 0.6481, 0.6582, 0.6683, 0.6784, 0.6885, 0.6986, 0.7087, 0.7188, 0.7289, 0.739, 0.7491, 0.7592, 0.7693, 0.7794, 0.7895, 0.7996, 0.8097, 0.8198, 0.8299, 0.84, 0.8501, 0.8602, 0.8703, 0.8804, 0.8905, 0.9006, 0.9107, 0.9209, 0.9311, 0.9412, 0.9514, 0.9616, 0.9718, 0.982, 0.9922)

plt.figure(figsize=(10,10))
plt.scatter(positions_uniques,last_energies[0:89], label='last band')
plt.scatter(first_curve_positions,first_curve_energies,s=0.5, c='red')
plt.ylabel("Energy (eV)")
plt.grid()
plt.legend()
plt.show()
print(last_energies[0:89])
print(positions_uniques)
```

```python
#NOW CONVERT VALUES TO REAL LATTICE VALUES, AND FROM eV to JOULES
real_positions = []
energies = []

for pos in positions_uniques:
    real_positions.append((pos*2*np.pi)/(3.073*10**10))
print(real_positions)

for en in last_energies[0:89]:
    energies.append(en*1.602*10**-19)
print(energies)
```

```python
#d = np.gradient(energies, real_positions)
plt.figure(figsize=(10,10))
plt.axvline(x=real_positions[47], label=f'min at {real_positions[47]}')
plt.scatter(real_positions, energies, label = 'band', s=3.0)
plt.ylabel("Energy (J)")
#plt.plot(real_positions, d)
#plt.plot(real_positions, second_derivative_energies)
plt.legend()
plt.grid()
plt.show()
```

```python
#ALL VALUES HAVE BEEN CONVERTED TO REAL LATTICE VALUES, AND FROM eV to JOULES:
print(real_positions)
print(energies)

energies = np.array(energies)
derivative_energies = np.zeros(energies.shape,np.float64)
derivative_energies[0:-1] = np.diff(energies)/np.diff(real_positions)
derivative_energies[-1] = (energies[-1] - energies[-2])/(real_positions[-1] - real_positions[-2])

second_derivative_energies = np.zeros(derivative_energies.shape,np.float64)
second_derivative_energies[0:-1] = np.diff(derivative_energies)/np.diff(real_positions)
second_derivative_energies[-1] = (derivative_energies[-1] - derivative_energies[-2])/(real_positions[-1] - real_positions[-2])

plt.figure(figsize=(10,10))
plt.axvline(x=real_positions[46], label=f'min at {real_positions[46]}')
plt.scatter(real_positions, energies, label = 'band', s=3.0)
plt.plot(real_positions, derivative_energies)
#plt.plot(real_positions, second_derivative_energies)
plt.ylabel("Energy (J)")
plt.legend()
plt.grid()
plt.show()
print(f"The value of the second derivative at the minimum is: {second_derivative_energies[46]}")
red_h = 1.054571817*10**(-34)
effective_mass = (red_h**2)/(second_derivative_energies[46])
print(effective_mass)
```
