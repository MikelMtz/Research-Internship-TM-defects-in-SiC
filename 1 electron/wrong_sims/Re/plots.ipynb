```python
import pandas as pd
import numpy as np

df = pd.read_csv('Output.csv', delim_whitespace=True)
Energy_values = []
Row_positions = []
positions_of_points = []

for t in range(159):
    n = df.iloc[[5+7*t]].index[0][0]
    n=int(n.replace("-",""))
    Row_positions.append(n)

print(Row_positions)
print(len(Row_positions))

for t in range(159):
    E = df.iloc[[5+7*t]].to_string()
    E = E[93:115]
    find = E.find(".")
    positions_of_points.append(find)
    Energy_values.append(E)

print(Energy_values)
print(positions_of_points)

string_energy = []
t=0
for value in Energy_values:
    new = ""
    find = positions_of_points[t]
    for k in range(11):
        new= new + (value[find-1+k])
    string_energy.append(new)
    t=t+1
print(string_energy)
Final_energy=[]
for energy in string_energy:
    Final_energy.append(float(energy))
print(Final_energy)
```
