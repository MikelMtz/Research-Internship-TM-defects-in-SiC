# 6H-SiC

```python
from pyxtal import pyxtal
from pyxtal.lattice import Lattice
import matplotlib.pyplot as plt
from ase.visualize import view

# Generate a random 6H-SiC structure with corrected lattice parameters
structure = pyxtal()

lat = Lattice.from_para(3.073, 3.073, 15.117, 90, 90, 120, ltype='hexagonal')

structure.from_random(3,186,['Si', 'C'],[6, 6],lattice = lat)
print(structure)

# Convert to ASE Atoms
ase_structure = structure.to_ase()

# Visualize the structure
view(ase_structure)
```

```python
ase_structure = structure.to_ase()

# Build the supercell (2x2x2 replication)
supercell = ase_structure * [4, 4, 4]

# Visualize the structure
view(supercell)
```

# 3C-SiC

```python
from pyxtal import pyxtal
from pyxtal.lattice import Lattice
import matplotlib.pyplot as plt
from ase.visualize import view

# Generate a random 6H-SiC structure with corrected lattice parameters
structure = pyxtal()

lat = Lattice.from_para(4.3596, 4.3596, 4.3596, 90, 90, 90, ltype='cubic')

structure.from_random(3,186,['Si', 'C'],[4, 4],lattice = lat)
print(structure)

# Convert to ASE Atoms
ase_structure = structure.to_ase()

# Visualize the structure
view(ase_structure)
```

```python
ase_structure = structure.to_ase()

# Build the supercell (2x2x2 replication)
supercell = ase_structure * [2, 2, 2]

# Visualize the structure
view(supercell)
```

# Convert to cartesian coordinates

```python
#Access lattice matrix
lattice_matrix = structure.lattice.matrix  # 3x3 matrix of lattice vectors

# Access fractional atomic positions and species
atomic_positions = [site.position for site in structure.atom_sites]  # Fractional positions
species = [site.specie for site in structure.atom_sites]  # Element types

# Convert fractional positions to Cartesian
cartesian_positions = [
    [sum(fractional[i] * lattice_matrix[i][j] for i in range(3)) for j in range(3)]
    for fractional in atomic_positions
]

# Print the Cartesian coordinates
print("Element | Cartesian Coordinates (Ã…)")
for elem, coords in zip(species, cartesian_positions):
    print(f"{elem:6} | {coords}")


fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
for coords in cartesian_positions:
    ax.scatter3D(coords[0],coords[1],coords[2])
plt.title("Td")
plt.legend()
plt.show()
```

# Trying to get the correct placement

```python
import math
import pandas as pd
bond_length = 1.89

coord_5 = (bond_length*math.sqrt(3)/2,0,bond_length*4/2)
coord_6 = (-bond_length*math.sqrt(3)/4,3*bond_length/4,bond_length*4/2)
coord_7 = (-bond_length*math.sqrt(3)/4,-3*bond_length/4,bond_length*4/2)

def atomic_c3v_2nd_neighbours(d):
    return (0,0,d), (d*math.sqrt(3)/2,0,-d/2), (-d*math.sqrt(3)/4,3*d/4,-d/2), (-d*math.sqrt(3)/4,-3*d/4,-d/2), coord_5, coord_6, coord_7, (d*math.sqrt(3)/2,0,-d*3/2), (-d*math.sqrt(3)/4,3*d/4,-d*3/2), (-d*math.sqrt(3)/4,-3*d/4,-d*3/2)

def distance(x1,x2):
    summa=0
    for i in range(len(x1)):
        summa += (x2[i]-x1[i])**2
    return math.sqrt(summa)

fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
t=0
for coords in atomic_c3v_2nd_neighbours(bond_length):
    ax.scatter3D(coords[0],coords[1],coords[2])
    if t<4:
        ax.plot((0,coords[0]),(0,coords[1]),(0,coords[2]))   
    t=t+1

for t in range(3):
    ax.plot((atomic_c3v_2nd_neighbours(bond_length)[0][0],atomic_c3v_2nd_neighbours(bond_length)[4+t][0]),(atomic_c3v_2nd_neighbours(bond_length)[0][1],atomic_c3v_2nd_neighbours(bond_length)[4+t][1]),(atomic_c3v_2nd_neighbours(bond_length)[0][2],atomic_c3v_2nd_neighbours(bond_length)[4+t][2]))    
    ax.plot((atomic_c3v_2nd_neighbours(bond_length)[1+t][0],atomic_c3v_2nd_neighbours(bond_length)[7+t][0]),(atomic_c3v_2nd_neighbours(bond_length)[1+t][1],atomic_c3v_2nd_neighbours(bond_length)[7+t][1]),(atomic_c3v_2nd_neighbours(bond_length)[1+t][2],atomic_c3v_2nd_neighbours(bond_length)[7+t][2]))    

ax.scatter3D(0,0,0, label=f'bond length is {distance(atomic_c3v_2nd_neighbours(bond_length)[3],(0,0,0))}')
plt.legend()
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.title("C3v at second neighbours")
plt.show()

def atomic_C3v_dataframe(d):
    data = {'1': [0,0,d],
           '2': [d*math.sqrt(3)/2,0,-d/2],
           '3': [-d*math.sqrt(3)/4,3*d/4,-d/2],
           '4': [-d*math.sqrt(3)/4,-3*d/4,-d/2],
           '5': coord_5,
           '6': coord_6,
           '7': coord_7,
           '8': [d*math.sqrt(3)/2,0,-d*3/2],
           '9': [-d*math.sqrt(3)/4,3*d/4,-d*3/2],
           '10': [-d*math.sqrt(3)/4,-3*d/4,-d*3/2]}
    df = pd.DataFrame(data)
    return df.transpose()
print(atomic_C3v_dataframe(bond_length))
```

```python
def atomic_td(d):
    return (-d/2,-d/2,d/2),  (-d/2,d/2,-d/2),  (d/2,-d/2,-d/2),  (d/2,d/2,d/2)

def atomic_td_2nd_neighbours_extension(d):
    return (d, d, 0), (0,d,-d), (d,0,-d), (-d,-d,0), (0,-d,d), (-d,0,d)

d = (bond_length*2)/math.sqrt(3)


fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
for coords in atomic_td(d):
    ax.scatter3D(coords[0],coords[1],coords[2])
    ax.plot((0,coords[0]),(0,coords[1]),(0,coords[2]))

for coords in atomic_td_2nd_neighbours_extension(d):
    ax.scatter3D(coords[0],coords[1],coords[2])

ax.plot((d,d/2), (d,d/2), (0,d/2))
ax.plot((0,-d/2), (d,d/2), (-d,-d/2))
ax.plot((d,d/2), (0,-d/2), (-d,-d/2))

ax.plot((-d,-d/2), (-d,-d/2), (0,d/2))
ax.plot((0,-d/2), (-d,-d/2), (d,d/2))
ax.plot((-d,-d/2), (0,-d/2), (d,d/2))


ax.scatter3D(0,0,0, label=f'bond length is {distance(atomic_td(d)[3],(0,0,0))}')
plt.title("Td")
plt.legend()
plt.show()
```
